```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
np.random.seed(42)

# Sample simulated social media posts
posts = [
    "I love this new phone! Amazing performance and battery life.",
    "Worst service ever. I'm so disappointed.",
    "The product quality is average. Could be better.",
    "Absolutely fantastic experience. Highly recommend!",
    "Terrible app. Keeps crashing.",
    "It's okay. Not great, not bad.",
    "Loved the packaging and quick delivery!",
    "Customer support was very rude.",
    "Iâ€™m happy with the purchase overall.",
    "The features are impressive and very useful!"
]

# Simulated sentiment labels
sentiments = ["Positive", "Negative", "Neutral", "Positive", "Negative",
              "Neutral", "Positive", "Negative", "Positive", "Positive"]

df = pd.DataFrame({'Post': posts, 'Sentiment': sentiments})
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sentiment', palette='Set2')
plt.title("Sentiment Distribution in Social Media Posts")
plt.xlabel("Sentiment")
plt.ylabel("Number of Posts")
plt.show()
```

```python
positive_text = " ".join(df[df['Sentiment'] == 'Positive']['Post'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud - Positive Sentiment")
plt.show()
```

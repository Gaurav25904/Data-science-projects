```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

np.random.seed(42)

data = pd.DataFrame({
    "Age": np.random.randint(18, 80, size=200),
    "Gender": np.random.choice(["Male", "Female", "Other"], size=200, p=[0.45, 0.45, 0.10])
})

plt.figure(figsize=(10, 5))
sns.histplot(data["Age"], bins=15, kde=True, color="skyblue")
plt.title("Age Distribution in Sample Population")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.grid(True)
plt.tight_layout()
plt.show()

gender_counts = data["Gender"].value_counts().reset_index()
gender_counts.columns = ["Gender", "Count"]

plt.figure(figsize=(8, 5))
sns.barplot(data=gender_counts, x="Gender", y="Count", palette="pastel")
plt.title("Gender Distribution in Sample Population")
plt.xlabel("Gender")
plt.ylabel("Number of People")
plt.tight_layout()
plt.show()
```

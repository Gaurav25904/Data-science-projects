```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
n = 500

data = pd.DataFrame({
    "Accident_ID": range(1, n + 1),
    "Datetime": pd.date_range(start="2025-07-01", periods=n, freq="H"),
    "Weather": np.random.choice(["Clear", "Rainy", "Fog", "Snow", "Storm"], n, p=[0.5, 0.25, 0.1, 0.1, 0.05]),
    "Road_Condition": np.random.choice(["Dry", "Wet", "Icy", "Snowy"], n, p=[0.6, 0.25, 0.1, 0.05]),
    "Severity": np.random.choice(["Minor", "Moderate", "Major", "Fatal"], n, p=[0.5, 0.3, 0.15, 0.05]),
    "Latitude": np.random.uniform(19.0, 19.3, n),
    "Longitude": np.random.uniform(72.8, 73.0, n)
})

# Extract time features
data["Hour"] = data["Datetime"].dt.hour
data["Day"] = data["Datetime"].dt.day
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(x='Hour', data=data, palette="viridis")
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Weather', hue='Severity', data=data)
plt.title("Weather vs. Accident Severity")
plt.xlabel("Weather Condition")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Road_Condition', data=data, palette="coolwarm")
plt.title("Road Condition Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Severity', data=data)
plt.title("Accident Hotspots by Severity")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```
